```python
# data set prep
dataset=pd.DataFrame(california.data,columns=california.feature_names)
dataset.head()
```

```python
dataset['Price']=california.target
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(dataset.corr(),annot=True)
```

```python
# dep and indep feature
X=dataset.iloc[:,:-1]#independent
Y=dataset.iloc[:,-1]#dependent
```

```python
from sklearn.model_selection import  train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=10)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train =scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
# model training
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
regression
```

```python
regression.fit(X_train,y_train)
```

```python
# slop or coefficients
regression.coef_
```

```python
regression.intercept_
```

```python
# pred for test data
y_pred=regression.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.metrics import r2_score

mae=mean_absolute_error(y_test,y_pred)
score=r2_score(y_test,y_pred)
print("mae",mae)
print("mae",mean_squared_error(y_test,y_pred))
print('r2 score',score)
```

```python
# adjusted r sq
adj_rsq= 1-(1-score)*len(y_test)/(len(y_test)-X_test.shape[1]-1)
adj_rsq
```

```python
# assumptions
plt.scatter(y_test,y_pred)
plt.xlabel("test truth data")
plt.ylabel("test predicted data")
```

```python
residuals = y_test-y_pred
```

```python
residuals
```

```python
sns.distplot(residuals,kde=True)
# to be cont distrib
```

```python
# plot with predictions and residual to be uniform distribution for good model
plt.scatter(y_pred,residuals)
```
